##  1. K means

### Cmo funciona?
- Paso 1: El algoritmo utiliza un proceso iterativo en el que se van ajustando los grupos para producir el resultado final. Para ejecutar el algoritmo deberemos pasar como entrada el conjunto de datos y un valor de K. El conjunto de datos sern las caractersticas o features para cada punto. Las posiciones iniciales de los K centroids sern asignadas de manera aleatoria de cualquier punto del conjunto de datos de entrada. Luego se itera en los siguientes dos pasos

- Paso 2: Asignar observaciones al centro de clster ms cercano.
    En este paso, cada "fila" de nuestro conjunto de datos se asigna al centroide ms cercano basado en la distancia cuadrada Euclideana (se peuden usar otras distancias).     
    
    Se utiliza la siguiente frmula (donde dist() es la distancia Euclideana standard):
    
    
    Hasta este punto, nuestros clusters estarin definidos de la siguiente manera:
    
- Paso 3: Actualizar el centroide
    En este paso los centroides de cada grupo son recalculados. Esto se hace tomando una media de todos los puntos asignados en el paso anterior.
    
    
    
El algoritmo itera entre estos ultimos dos pasos hasta cumplir un criterio de detencin:
- si no hay cambios en los puntos asignados a los grupos,
- o si la suma de las distancias se minimiza,
- o se alcanza un nmero mximo de iteraciones.

El algoritmo converge a un resultado que puede ser el ptimo local, por lo que ser conveniente volver a ejecutar ms de una vez con puntos iniciales aleatorios para confirmar si hay una salida mejor.


### Elegir el valor del nmero de clusters (K)
K means funciona definiendo apiori un valor de K. Para encontrar el nmero de clusters en los datos, deberemos ejecutar el algoritmo para un rango de valores K, ver los resultados y comparar caractersticas de los grupos obtenidos. En general no hay un modo exacto de determinar el valor K, pero se puede estimar con aceptable precisin siguiendo la siguiente tcnica:

Una de las mtricas usada para comparar resultados es la distancia media entre los puntos de datos y su centroid. Como el valor de la media disminuir a medida de aumentemos el valor de K, deberemos utilizar la distancia media al centroide en funcin de K y entontrar el "punto codo", donde el margen del ratio de descenso disminuye. Aqu vemos una grfica a modo de ejemplo:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
 
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
#plt.rcParams['figure.figsize'] = (16, 9)
#plt.style.use('ggplot')
```

### ... Seguiremos trabajando con nuestra data de Attrition

```python
df_clientes = pd.read_excel('../data/train_clientes.xlsx')
```

```python
df_clientes.index = df_clientes['ID_CORRELATIVO']
df_clientes.drop(columns = ['ID_CORRELATIVO','CODMES'], inplace = True)
```

```python
df_clientes.head()
```

```python
# Creamos variables dummies
df_clientes_km = pd.get_dummies(df_clientes)

# Imputamos los missig por la media
df_clientes_km.fillna(np.mean(df_clientes_km['EDAD']), inplace=True)
df_clientes_km.fillna(np.mean(df_clientes_km['ANTIGUEDAD']), inplace = True)
```

```python
#dataframe = pd.read_csv('../data/analisis.csv')
#|dataframe.head()
```

```python
# Creamos las matrices para ejecutar nuestro modelo
X = np.array(df_clientes_km.loc[:, df_clientes_km.columns != 'ATTRITION'])
y = np.array(df_clientes_km['ATTRITION'])
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)# Si se quiere fijar la semilla random_state=0
```

```python
import seaborn as sns; sns.set()
from scipy.spatial.distance import cdist

distorcion = []
K = range(1,20)
for k in K:
    kmeans = KMeans(n_clusters=k).fit(X_train)
    kmeans.fit(X_train)
    distorcion.append(sum(np.min(cdist(X_train, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X_train.shape[0])

# Plot the elbow
plt.figure(figsize=(8,8)) #definir el tamao de la imagen
plt.plot(K, distorcion, 'bx-')
plt.xlabel('Nmero de clusters')
plt.ylabel('Distorcin')
plt.title('El mtodo del Codo muestra el k ptimo')
plt.show()
```

### Ejecutamos k means

```python
# Entrenamos el modelo
kmeans = KMeans(n_clusters=4).fit(X_train) # en este caso 4 clusters
centroids = kmeans.cluster_centers_
```

```python
# Predecimos los clusters
labels = kmeans.predict(X_train)
```

```python
# Creamos un dataframe con nuestras variables y las etiquetas de los cluster
df_km = pd.DataFrame(X_train, columns = df_clientes_km.loc[:, df_clientes_km.columns != 'ATTRITION'].columns)
df_km['clusters'] = labels
```

```python
df_km.groupby('clusters').agg({'EDAD':'mean','ANTIGUEDAD':'mean','clusters':'count'})
```

```python
sb.pairplot(df_km, hue='clusters',height=4,vars=['EDAD','ANTIGUEDAD'],kind='scatter')
```

### Ojo: hay que procuprar estandarizar nuestros datos

```python
X_train_std = StandardScaler().fit_transform(X_train)

# Entrenar el modelo
kmeans = KMeans(n_clusters=4).fit(X_train_std)

# Predecimos los clusters
labels = kmeans.predict(X_train_std)

df_km = pd.DataFrame(X_train, columns = df_clientes_km.loc[:, df_clientes_km.columns != 'ATTRITION'].columns)
df_km['clusters'] = labels
```

```python
sb.pairplot(df_km, hue='clusters',height=4,vars=["EDAD","ANTIGUEDAD"],kind='scatter')
```

Buscaremos los usuarios que estn ms cerca a los centroids de cada cluster.
Podramos decir que tienen las caractersticas que representan a cada cluster

```python
# Veamos al representante de cada cluster (el usuario cercano a al centroide de cada cluster)
usr_rep, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, X_train_std)
usr_rep
```

# 2. K modes

- El algoritmo de agrupacin de los K modes fue publicado por primera vez en 1997 (Huang, 1997b). Se public un enfoque equivalente independientemente en (Chaturvedi, Green, Carroll, & Foods, 2001). La relacin de los dos mtodos de k-modes se describe en (Huang & Ng, 2003).

- Es una extensin del algoritmo estandar de K means para clustering con datos categricos.
- Las principales modificaciones respecto a k-means incluyen la funcin de distancia, la representacin del centro del cluster y el proceso iterativo de agrupamiento.

### Funcin de la distancia
Para calcular la distancia (o disimilitud) entre dos objetos X e Y descrito por *m* atributos categricos, la funcin de distancia en k-modes se define como:


Donde, xj e yj son los valores del atributo j en X e Y. 
Esta funcin se conoce a menudo como la medida de disimilitud de coincidencia simple o la distancia de Hemming. Miestras mayor sea la diferencia de los valores categricos entre X e Y, mayor ser la dimimilitud entre los dos objetos.

En K modes, los centros del clster estn representados por los vectores de las modas de atributos categricos. En Estadstica, la moda de un conjunto de valores es el valor ms frecuente.

## Cmo funciona?

- Paso 1: Selecciona aletoriamente k unicos objetos como el nmero inicial de centros del cluster (modas)
- Paso 2: Calcula la distancia entre cada valor y la moda del cluster, asigna el objeto al cluster cuyo centro tenga la distancia ms corta al objeto; repita este paso hasta que todos los objetos sean asignados a clusters.
- Paso 3: Seleccione una nueva moda para cada clster y comprelo con la moda anterior. Si es diferente, regrese al Paso 2; de lo contrario detngase.
  Este proceso de clustering minimiza la siguiente funcin objetivo K modes:
  
  
Donde U = [ui,l] es una particion de matriz nk, Z = {Z1,Z2,  , Zk} es un conjunto de vector de modas y la funcin distancia d(.,.) lo definimos anteriormente. Dado que K modes es esencialmente lo mismo que K means, el algoritmo de K modes es eficiente en la agrupacin de grandes datos categricos y tambin produce localmente resultados mnimos de agrupacin.

```python
import pandas as pd
from kmodes.kmodes import KModes
```

```python
df_kmod = df_clientes.select_dtypes(include='object').fillna('otro')
columns_cat = df_kmod.columns

for col in columns_cat:
    # Cambiamos el typo de datos a categrica
    df_kmod[col] = df_kmod[col].astype('category')
    col_new = col + '_ENC'
    df_kmod[col_new] = df_kmod[col].cat.codes
```

```python
# instanciar k-modes object - 5 clusters
kmodes = KModes(n_clusters = 5 , init='Cao', verbose=1)

# modelando
kmodes.fit(df_kmod.filter(like='_ENC'))
```

```python
# centroides de los clusters
print(kmodes.cluster_centroids_)
```

```python
# centroides de los clusters
print(kmodes.cluster_centroids_)
```

```python
# Agregamos la etiqueta de los clusters
df_kmod['kmodes'] = kmodes.labels_
```

```python
df_kmod.groupby('kmodes').size()
```

```python
usr_rep, _ = pairwise_distances_argmin_min(kmodes.cluster_centroids_,
                                           df_kmod.filter(like='_ENC'), 
                                           metric = 'jaccard')
usr_rep
```

# 2. K prototypes

El algoritmo de K prototypes puedes ser descrito como:
1. Selececiona un k inicial de prototypes de un data set X, uno por cada cluster
2. Asigne cada objeto en X a un cluster cuyo prototype sea el ms cercano segn la siguiente ecuacin:


Actualiza el prototype del cluster despues de cada asignacin.
3. Despus de que todos los objetos se hayan asignado a un clster, vuelve a generar la similaridad de objetos a los prototypes actuales. 
Si se encuentra un objeto de modo que su prototipo ms cercano pertenezca a otro clster en lugar de al actual, reasigne el objeto a ese clster y actualice los prototipos de ambos clsteres.

4. Repita el paso 3 hasta que ningn objeto haya cambiado de cluster despus un ciclo completo de X

```python
from kmodes.kprototypes import KPrototypes
```

```python
# Preparamos la data

# Imputamos el promedio para las variables numricas
df_clientes['EDAD'] = df_clientes['EDAD'].fillna(np.mean(df_clientes['EDAD']))
df_clientes['ANTIGUEDAD'] = df_clientes['ANTIGUEDAD'].fillna(np.mean(df_clientes['ANTIGUEDAD']))

# Creamos una categora para los nulos, recategorizamos los null por 'missing'
df_clientes.fillna('missing',inplace=True)

df_kpt = df_clientes
#df_kpt.drop(columns='ATTRITION',inplace=True)
```

```python
df_kpt.head()
```

```python
# En este caso, slo entrenamos con el 20% de los datos
porc_sample = 0.2
df_kpt_train = df_kpt.sample(frac = porc_sample, replace=False)
df_kpt_test = df_kpt.loc[~df_kpt.index.isin(df_kpt_train.index)]
```

```python
print('Num Train: ',len(df_kpt_train))
print('Num Test: ',len(df_kpt_test))
```

```python
# Identificamos los indices de las columnas categricas
cols_cat = df_kpt.select_dtypes(include='object').columns
cols_cat_index = [df_kpt.columns.get_loc(c) for c in df_kpt.columns if c in cols_cat]
cols_cat_index
```

```python
# Iniciamos con 5 clusters
kprototypes = KPrototypes(n_clusters = 5, max_iter=10)
```

```python
# Entrenamiento
kprototypes.fit(df_kpt_train.values, categorical = cols_cat_index)
```

```python
df_kpt_train['kprototypes'] = kprototypes.labels_
```

```python
df_kpt_train.groupby('kprototypes').size()
```
