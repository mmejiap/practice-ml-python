```python
def number_assistance(dataset,patient_ID,registration_date):
    patient_data = dataset[(dataset['Target'] == 1) & (dataset['Patient_ID'] == patient_ID) & (dataset['Camp_End_Date']<registration_date)].shape[0]
    #patient_data = patient_data[patient_data['Patient_ID'] == patient_ID | patient_data['Camp_End_Date']<registration_date]
    #patient_data = patient_data[patient_data['Camp_End_Date']<registration_date]
    return patient_data

def number_register(dataset,patient_ID,registration_date):
    #patient_data = dataset[dataset['Target'] == 1]
    patient_data = dataset[(dataset['Patient_ID'] == patient_ID) & (dataset['Camp_End_Date']<registration_date)].shape[0]
    #patient_data = patient_data[patient_data['Camp_End_Date']<registration_date]
    return patient_data
```

```python
def discretize_by_density(dataset, variable, target, binds):
        ok = True
        while ok and binds > 0:
            try:
                discretize = pd.qcut(dataset[variable], binds)
                groups = dataset.groupby(discretize)                
                ok = False
            except ValueError:
                ok = True
                binds -=1
        if binds == 0:
            print ('Error con numero de binds')            
        elif binds < 3:
            print ('Pocos binds')
        else:
            efective = groups.mean()[target]
            x = range(binds)
            my_xticks = list(efective.index)
            
            plt.plot(range(binds), efective, '--o', range(binds), binds*[dataset[target].mean()], '-')
            plt.legend([variable,'umbral'], loc='upper left', prop={'size':8})
            plt.xticks(x, my_xticks, rotation=30, ha='right', size=8)
            plt.title('Efectividad')
            
            for i in range(binds):
                plt.text(i, efective.values[i], discretize.value_counts()[i])
            plt.show()
            
            return dataset
```
